**a.** Dibuje las gr√°ficas para $ùë¶ = ùë•$ y $ùë¶ = tan ùë•.$


```python
import numpy as np
import matplotlib.pyplot as plt

# Definici√≥n de valores de x entre 0 y pi/2 (no se incluye pi/2 para evitar la as√≠ntota)
x_vals = np.linspace(0.01, np.pi/2 - 0.01, 400)
y1 = x_vals
y2 = np.tan(x_vals)

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y1, label='y = x', color='blue')
plt.plot(x_vals, y2, label='y = tan(x)', color='green')
plt.title("Gr√°ficas de y = x y y = tan(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.axvline(np.pi/2, color='red', linestyle='--', label='x = œÄ/2 (As√≠ntota)')
plt.ylim(0, 5)
plt.grid(True)
plt.legend()
plt.show()
```
